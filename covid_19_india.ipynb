```
def lookup_tweets(tweet_ids, Api):
    tweets = []
    tweet_count = len(tweet_ids)
    for i in range(1 + tweet_count//100):
        full_tweets = []
        input = tweet_ids[i*100 : min(100*(i+1), tweet_count)]
        try:
            full_tweets.extend(Api.statuses_lookup(id_ = input))
            tweets.append(full_tweets) 
        except:
            print("error")
    
    return tweets
```

```
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def preprocess(tweet):
    tweet = re.sub(r'\&w*;', '', tweet) #removes html specific entities
    tweet = re.sub('@[^\s]+', '', tweet) #removes @
    tweet = re.sub(r'\$\w*', '', tweet) #removes tickers
    tweet = tweet.lower()
    tweet = re.sub(r'https?:\/\/.*\/\w*', '', tweet) #removes hyperlinks
    tweet = re.sub(r'\#\w', '', tweet) #removes hashtags
    for word in tweet.split():
        if word in stopwords.words('english'):
            tweet.replace(word, '')
    tweet = re.sub(r'\s\s+', '', tweet) #removes whitespace
    tweet = tweet.lstrip(' ')
    tweet = ''.join(c for c in tweet if c <= "\uFFFF")
    return tweet
```

```
final_df['tweet'] = final_df['tweet'].apply(lambda x : preprocess(x))
```
